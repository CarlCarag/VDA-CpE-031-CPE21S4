```python
#Question 1. What is the average BMI for 20 Males and 20 Females surveyed?
import matplotlib.pyplot as plt

categories = ['Male','Female']
values = [29.94,30.21]

plt.bar(categories,values)
plt.title("BMI of Male vs BMI of Female")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

Description: This means that the results of the BMI among the 9 males in the 20 surveyed people in Alabama State has 30.21% which means that majority of the males surveyed are in obese category. While the 11 Females in the 20 surveyed people in Alabama State has a total of 29.94% which means that half of the population is in the normal category. Proper BMI results could help prevent the occuring of illnessess including cardiovascular diseases.

```python
#Question 2. How many Males and Females surveyed who had covid?
import matplotlib.pyplot as plt

labels = ['Male (4 Males)', 'Female (3 Females)']
sizes = [60,40]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```
