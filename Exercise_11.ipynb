```python
#Sample Problem
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Generate Data
x = np.array([0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40 ])
y = np.array([1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23])
#Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)
#Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title('Sample Problem')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Regression Line', 'Data Points'])
plt.show()
```

Interpretation:
This graph shows 2 data sets, the two data sets shows have a correlation result of r = -0.9475, which shows a strong negative correlation
