```python
#In total of 10 surveyed male and female, Alabama State, What are the common categories of general health.
import matplotlib.pyplot as plt
#Figure 1

GeneralHealth = ['Poor','Fair','Good','Very Good', 'Excellent']
TotalNumberofPeople = [1,2,3,3,1]

plt.bar(GeneralHealth, TotalNumberofPeople)
plt.title("Common General Health Category")
plt.xlabel("General Health")
plt.ylabel("Total Number of People")
plt.show()
```

Description: Among the 10 surveyed people in Alabama there are 3 persons who stated that they have a very good health. While following it with the same number of persons which is equal to 3 that states they good health. 2 persons stated that they have a fair health. The least frequency obtaining 1 person who states that they have an excellent health, and 1 person who stated that they have a poor health.

```python
#In total of 10 surveyed male and female, Alabama State, What are the number of people with average hours of sleep.
#Figure 2
import matplotlib.pyplot as plt

labels = ['5 hours of sleep(5%)', '7 hours of sleep(80%)', '6 hours of sleep(5%)','8 hours of sleep(5%)','9 hours of sleep(5%)' ]
sizes = [5,80,5,5,5]

plt.pie(sizes, labels=labels)
plt.title("Among 10 people in Alabama highest frequency of hours of sleep")
plt.show()
```

Description: Among the 10 surveyed people in Alabama, majority of the number of hours of sleep is 7 hours, while the remaining sleep hours fall in the same percentages as 5 hrs, 9 hrs, 8, hrs, and 6 hours of sleep.
