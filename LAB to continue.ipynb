```python

#A. Read each device's date of release and sales. Show it using a line plot


plt.figure(figsize=(10,9))
x = df['Date of Release']
y = df['Price (USD)']
plt.plot(x,y,)
plt.title("Line Plot Example")
plt.xlabel("'Date of Release")
plt.ylabel("Price (USD)")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,9))
x = df['Date of Release']
y = df['Price (USD)']
#B. Linestyle = dotted, Line Color = red,
plt.plot(x,y, linestyle='dotted', color='red', marker='o' )
plt.title("Line Plot Example")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,9))
x = df['Date of Release']
y = df['Price (USD)']
#B. Linestyle = dotted, Line Color = red,
plt.plot(x,y, linestyle='dotted', color='red', marker='o' )
plt.legend(["Price (USD)", "Release Date"], loc="lower right")
plt.title("Line Plot Example")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(rotation=60)
plt.show()
```
